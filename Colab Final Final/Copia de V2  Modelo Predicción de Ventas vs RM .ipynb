```python
# Detección de Outliers
# Método IQR para detectar outliers
def detectar_outliers_iqr(df, columna):
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    outliers = df[(df[columna] < limite_inferior) |
                  (df[columna] > limite_superior)]
    return len(outliers)

#   Aplicar a variables numéricas
columnas_numericas = df.select_dtypes(include=[np.number]).columns
for col in columnas_numericas:
    n_outliers = detectar_outliers_iqr(df, col)
    print(f"{col}: {n_outliers} outliers")

#Inversión CAPEX: 108 outliers → Algunos locales invierten mucho más (locales premium)
#Marketing: 128 outliers → Estrategias agresivas de marketing
#Alquiler: 101 outliers → Locales en ubicaciones premium (shopping de lujo)
#Ticket Promedio: 40 outliers → Negocios de alta gama (joyerías, tecnología)
#Ventas: 66 outliers → Los locales top performers
#No los eliminaremos - ya que son casos de negocio legítimos (ej: una joyería en shopping ABC1 tendrá ventas muy superiores al promedio).
```

```python
# Variables a excluir
excluir = ['ID_Local',
           'Margen_EBITDA_Pct',           # data leakage - resultado operativo
           'Ticket_Promedio_USD',          # data leakage - se conoce después de operar
           'Venta_Online_Pct',             # data leakage - se conoce después de operar
           'Ventas_Anuales_Totales_USD']   # Target variable

# NOTA: Rubro y Tipo_Ubicacion ya NO están en df_procesado como object
# porque get_dummies las convirtió a columnas dummies automáticamente
# por eso no hace falta excluirlas acá

# Seleccionar features
features = [col for col in df_procesado.columns if col not in excluir]

# Definir X e y
X = df_procesado[features]
y = df_procesado['Ventas_Anuales_Totales_USD']

print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
print(f"\nFeatures incluidas ({len(features)}):")
for f in features:
    print(f"  - {f}")
```

```python
# split (sobre X sin escalar)
X_temp, X_test, y_temp, y_test = train_test_split(
    X, y, test_size=0.15, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(
    X_temp, y_temp, test_size=0.176, random_state=42)

print(f"Train: {len(X_train)} | Val: {len(X_val)} | Test: {len(X_test)}")

# escalar, aprendiendo SOLO del train
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)  # aprende solo del train
X_val_scaled   = scaler.transform(X_val)         # solo aplica la transformación
X_test_scaled  = scaler.transform(X_test)        # solo aplica la transformación

print("■ Datos normalizados correctamente (fit solo en train)")
```

```python
# Entrenar KNN
# Buscar mejor K
from sklearn.model_selection import cross_val_score

k_range = range(1, 31)
val_rmse = []

for k in k_range:
    knn = KNeighborsRegressor(n_neighbors=k, weights='distance')
    scores = cross_val_score(knn, X_train_scaled, y_train,   # ← X_train_scaled
                             cv=5, scoring='neg_root_mean_squared_error')
    val_rmse.append(-scores.mean())

mejor_k    = list(k_range)[np.argmin(val_rmse)]
mejor_rmse = min(val_rmse)
print(f"Mejor K: {mejor_k}  →  RMSE CV: ${mejor_rmse:,.0f}")

# Gráfico — solo validación, que es lo que importa
plt.figure(figsize=(10, 5))
plt.plot(k_range, val_rmse, 'b-o', markersize=5, label='RMSE CV (5-fold)')
plt.axvline(x=mejor_k, color='red', linestyle='--', linewidth=2,
            label=f'Mejor K = {mejor_k}')
plt.xlabel('Valor de K', fontsize=12)
plt.ylabel('RMSE (USD)', fontsize=12)
plt.title('Selección del K óptimo en KNN', fontsize=14, fontweight='bold')
plt.legend()
plt.xticks(range(1, 31))
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
# Entrenar modelo KNN final con el mejor K
modelo_knn = KNeighborsRegressor(n_neighbors=mejor_k, weights='distance')
modelo_knn.fit(X_train_scaled, y_train)   # ← escalado
print(f"✅ Modelo KNN entrenado con K={mejor_k}")
```

```python
# -*- coding: utf-8 -*-
#EXPERIMENTO: IMPACTO DE OUTLIERS EN KNN vs RANDOM FOREST.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# =============================================================================
# CONFIGURACIÓN INICIAL
# =============================================================================
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
import warnings
warnings.filterwarnings('ignore')

# =============================================================================
# CARGA DE DATOS (ajusta la ruta según corresponda)
# =============================================================================
path = "/content/drive/MyDrive/KNN - Hackaton/dataset_locales_comerciales_3000_AMPLIADO.csv"
df = pd.read_csv(path)
print("Dataset original cargado:", df.shape)

# =============================================================================
# DEFINICIÓN DE VARIABLES A UTILIZAR (EVITANDO DATA LEAKAGE)
# =============================================================================
# Variables que NO se conocen antes de abrir el local (excluir)
leakage_vars = ['Ticket_Promedio_USD', 'Venta_Online_Pct', 'Margen_EBITDA_Pct']

# Variables a usar en el modelo (todas las demás numéricas + categóricas procesadas)
# Incluimos 'Alquiler_Mensual_USD' a pesar de la endogeneidad para mantener comparabilidad,
# pero en la versión final se recomienda excluirla o instrumentarla.
cols_numericas = ['Superficie_m2', 'Metros_Vidriera', 'Inversion_CAPEX_USD',
                  'Antiguedad_Anios', 'Inversion_Marketing_Mensual_USD',
                  'Alquiler_Mensual_USD', 'Transito_Peatonal_Pico',
                  'Puntos_Interes_Proximos', 'Distancia_Ancla_Metros',
                  'Ubicacion_Esquina', 'Densidad_Poblacional_Hab_Km2',
                  'Densidad_Competencia']

# Variables categóricas a codificar
categoricas = ['Rubro', 'Nivel_Socioeconomico', 'Tipo_Ubicacion', 'Posicionamiento_Precio']

# Variable objetivo
target = 'Ventas_Anuales_Totales_USD'

# =============================================================================
# FUNCIÓN PARA DETECTAR OUTLIERS (IQR)
# =============================================================================
def detectar_outliers_iqr(df, columnas, factor=1.5):
    """
    Devuelve una máscara booleana con True para las filas que son outliers
    en al menos una de las columnas especificadas.
    """
    mascara = pd.Series(False, index=df.index)
    for col in columnas:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lim_inf = Q1 - factor * IQR
        lim_sup = Q3 + factor * IQR
        mascara = mascara | ((df[col] < lim_inf) | (df[col] > lim_sup))
    return mascara

# Seleccionar variables numéricas donde tiene sentido buscar outliers
vars_outlier = ['Inversion_CAPEX_USD', 'Inversion_Marketing_Mensual_USD',
                'Alquiler_Mensual_USD', 'Ventas_Anuales_Totales_USD',
                ]

mascara_outliers = detectar_outliers_iqr(df, vars_outlier)

print("\n" + "="*80)
print("DETECCIÓN DE OUTLIERS")
print("="*80)
print(f"Total de registros: {len(df)}")
print(f"Registros con outliers: {mascara_outliers.sum()} ({mascara_outliers.sum()/len(df)*100:.1f}%)")
print("\nDistribución por variable:")
for var in vars_outlier:
    q1 = df[var].quantile(0.25)
    q3 = df[var].quantile(0.75)
    iqr = q3 - q1
    lim_inf = q1 - 1.5*iqr
    lim_sup = q3 + 1.5*iqr
    out = ((df[var] < lim_inf) | (df[var] > lim_sup)).sum()
    print(f"  {var}: {out} outliers ({out/len(df)*100:.1f}%)")

# =============================================================================
# FUNCIÓN PARA PREPROCESAR UN DATAFRAME (encoding + feature selection)
# =============================================================================
def preparar_dataset(df_input):
    df_proc = df_input.copy()

    categorical_cols = df_proc.select_dtypes(include=['object']).columns
    df_proc = pd.get_dummies(df_proc, columns=categorical_cols, drop_first=True)

    excluir = ['ID_Local',
               'Margen_EBITDA_Pct',
               'Ticket_Promedio_USD',
               'Venta_Online_Pct',
               'Ventas_Anuales_Totales_USD']

    features = [col for col in df_proc.columns if col not in excluir]

    X = df_proc[features].copy()
    y = df_proc[target].copy()
    return X, y, features

# =============================================================================
# PREPARAR AMBOS DATASETS (ORIGINAL Y SIN OUTLIERS)
# =============================================================================
print("\n" + "="*80)
print("PREPARACIÓN DE DATASETS")
print("="*80)

# Dataset original
X_orig, y_orig, features = preparar_dataset(df)
print(f"Original: {X_orig.shape}")

# Dataset sin outliers
df_clean = df[~mascara_outliers].copy()
X_clean, y_clean, _ = preparar_dataset(df_clean)
print(f"Sin outliers: {X_clean.shape}")

# =============================================================================
# FUNCIÓN PARA ENTRENAR Y EVALUAR MODELOS
# =============================================================================
def entrenar_modelos(X, y, nombre):
    """
    Entrena KNN (con búsqueda de k óptimo) y Random Forest.
    Retorna diccionario con métricas y los modelos entrenados (opcional).
    """
    # --- Train/Validation/Test split (misma semilla para ambos) ---
    X_temp, X_test, y_temp, y_test = train_test_split(
        X, y, test_size=0.15, random_state=42)
    X_train, X_val, y_train, y_val = train_test_split(
        X_temp, y_temp, test_size=0.176, random_state=42)  # 0.176 de 0.85 = 0.15

    print(f"\n{nombre} -> Train: {X_train.shape}, Val: {X_val.shape}, Test: {X_test.shape}")

    # --- Escalado (solo para KNN) ---
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_val_scaled = scaler.transform(X_val)
    X_test_scaled = scaler.transform(X_test)

    # --- KNN: búsqueda del mejor k ---
    k_range = range(1, 31)
    val_rmse = []
    for k in k_range:
        knn = KNeighborsRegressor(n_neighbors=k, weights='distance')
        scores = cross_val_score(knn, X_train_scaled, y_train,
                                 cv=5, scoring='neg_root_mean_squared_error')
        val_rmse.append(-scores.mean())
    best_k = list(k_range)[np.argmin(val_rmse)]

    # Entrenar KNN final con mejor k
    knn_final = KNeighborsRegressor(n_neighbors=best_k, weights='distance')
    knn_final.fit(X_train_scaled, y_train)
    y_pred_knn = knn_final.predict(X_test_scaled)

    # Métricas KNN
    mae_knn = mean_absolute_error(y_test, y_pred_knn)
    rmse_knn = np.sqrt(mean_squared_error(y_test, y_pred_knn))
    r2_knn = r2_score(y_test, y_pred_knn)

    # --- Random Forest (sin escalado) ---
    rf = RandomForestRegressor(
        n_estimators=150,
        max_depth=15,
        min_samples_split=2,
        random_state=42,
        n_jobs=-1
    )
    rf.fit(X_train, y_train)  # RF no necesita escalado
    y_pred_rf = rf.predict(X_test)

    mae_rf = mean_absolute_error(y_test, y_pred_rf)
    rmse_rf = np.sqrt(mean_squared_error(y_test, y_pred_rf))
    r2_rf = r2_score(y_test, y_pred_rf)

    # Mostrar resultados
    print(f"\n  KNN (k={best_k}): R²={r2_knn:.4f}, MAE=${mae_knn:,.0f}, RMSE=${rmse_knn:,.0f}")
    print(f"  RF:             R²={r2_rf:.4f}, MAE=${mae_rf:,.0f}, RMSE=${rmse_rf:,.0f}")

    return {
        'knn': {'r2': r2_knn, 'mae': mae_knn, 'rmse': rmse_knn, 'best_k': best_k},
        'rf':  {'r2': r2_rf,  'mae': mae_rf,  'rmse': rmse_rf}
    }

# =============================================================================
# EJECUTAR EXPERIMENTO
# =============================================================================
print("\n" + "="*80)
print("ENTRENAMIENTO CON DATASET ORIGINAL")
print("="*80)
res_orig = entrenar_modelos(X_orig, y_orig, "ORIGINAL")

print("\n" + "="*80)
print("ENTRENAMIENTO CON DATASET SIN OUTLIERS")
print("="*80)
res_clean = entrenar_modelos(X_clean, y_clean, "SIN OUTLIERS")

# =============================================================================
# COMPARACIÓN DE RESULTADOS
# =============================================================================
print("\n" + "="*80)
print("COMPARACIÓN DE RESULTADOS")
print("="*80)

comparacion = pd.DataFrame({
    'Modelo': ['KNN', 'KNN', 'KNN', 'RF', 'RF', 'RF'],
    'Métrica': ['R²', 'MAE', 'RMSE', 'R²', 'MAE', 'RMSE'],
    'Con Outliers': [
        res_orig['knn']['r2'], res_orig['knn']['mae'], res_orig['knn']['rmse'],
        res_orig['rf']['r2'],  res_orig['rf']['mae'],  res_orig['rf']['rmse']
    ],
    'Sin Outliers': [
        res_clean['knn']['r2'], res_clean['knn']['mae'], res_clean['knn']['rmse'],
        res_clean['rf']['r2'],  res_clean['rf']['mae'],  res_clean['rf']['rmse']
    ]
})
comparacion['Cambio %'] = ((comparacion['Sin Outliers'] - comparacion['Con Outliers'])
                           / comparacion['Con Outliers'] * 100)

print("\n📊 Tabla comparativa:")
print(comparacion.to_string(index=False, float_format=lambda x: f'{x:,.3f}' if abs(x)>100 else f'{x:.3f}'))

# =============================================================================
# GRÁFICOS MEJORADOS
# =============================================================================
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.patch.set_facecolor('#f5f5f5')

# Colores
color_out = '#6baed6'   # azul claro
color_clean = '#08519c' # azul oscuro
bar_width = 0.35

# Posiciones de las barras
indices = np.arange(2)  # dos grupos: KNN y RF

# 1. R²
ax = axes[0]
ax.bar(indices - bar_width/2,
       [res_orig['knn']['r2'], res_orig['rf']['r2']],
       bar_width, label='Con Outliers', color=color_out, edgecolor='black')
ax.bar(indices + bar_width/2,
       [res_clean['knn']['r2'], res_clean['rf']['r2']],
       bar_width, label='Sin Outliers', color=color_clean, edgecolor='black')
ax.set_xticks(indices)
ax.set_xticklabels(['KNN', 'Random Forest'])
ax.set_ylabel('R² (mayor es mejor)')
ax.set_title('Comparación de R²')
ax.legend(loc='lower right')
ax.grid(axis='y', alpha=0.3)
# Valores en las barras
for i, (orig, clean) in enumerate(zip([res_orig['knn']['r2'], res_orig['rf']['r2']],
                                        [res_clean['knn']['r2'], res_clean['rf']['r2']])):
    ax.text(i - bar_width/2, orig + 0.01, f'{orig:.3f}', ha='center', va='bottom', fontsize=9)
    ax.text(i + bar_width/2, clean + 0.01, f'{clean:.3f}', ha='center', va='bottom', fontsize=9)

# 2. MAE
ax = axes[1]
ax.bar(indices - bar_width/2,
       [res_orig['knn']['mae']/1000, res_orig['rf']['mae']/1000],
       bar_width, label='Con Outliers', color=color_out, edgecolor='black')
ax.bar(indices + bar_width/2,
       [res_clean['knn']['mae']/1000, res_clean['rf']['mae']/1000],
       bar_width, label='Sin Outliers', color=color_clean, edgecolor='black')
ax.set_xticks(indices)
ax.set_xticklabels(['KNN', 'Random Forest'])
ax.set_ylabel('MAE (miles USD)')
ax.set_title('MAE (menor es mejor)')
ax.legend(loc='upper right')
ax.grid(axis='y', alpha=0.3)
for i, (orig, clean) in enumerate(zip([res_orig['knn']['mae']/1000, res_orig['rf']['mae']/1000],
                                        [res_clean['knn']['mae']/1000, res_clean['rf']['mae']/1000])):
    ax.text(i - bar_width/2, orig + 1, f'{orig:.0f}k', ha='center', va='bottom', fontsize=9)
    ax.text(i + bar_width/2, clean + 1, f'{clean:.0f}k', ha='center', va='bottom', fontsize=9)

# 3. RMSE
ax = axes[2]
ax.bar(indices - bar_width/2,
       [res_orig['knn']['rmse']/1000, res_orig['rf']['rmse']/1000],
       bar_width, label='Con Outliers', color=color_out, edgecolor='black')
ax.bar(indices + bar_width/2,
       [res_clean['knn']['rmse']/1000, res_clean['rf']['rmse']/1000],
       bar_width, label='Sin Outliers', color=color_clean, edgecolor='black')
ax.set_xticks(indices)
ax.set_xticklabels(['KNN', 'Random Forest'])
ax.set_ylabel('RMSE (miles USD)')
ax.set_title('RMSE (menor es mejor)')
ax.legend(loc='upper right')
ax.grid(axis='y', alpha=0.3)
for i, (orig, clean) in enumerate(zip([res_orig['knn']['rmse']/1000, res_orig['rf']['rmse']/1000],
                                        [res_clean['knn']['rmse']/1000, res_clean['rf']['rmse']/1000])):
    ax.text(i - bar_width/2, orig + 1, f'{orig:.0f}k', ha='center', va='bottom', fontsize=9)
    ax.text(i + bar_width/2, clean + 1, f'{clean:.0f}k', ha='center', va='bottom', fontsize=9)

plt.suptitle('Impacto de la eliminación de outliers en KNN y Random Forest', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('outlier_experiment.png', dpi=150, bbox_inches='tight', facecolor='#f5f5f5')
plt.show()

# =============================================================================
# CONCLUSIONES AUTOMÁTICAS
# =============================================================================
print("\n" + "="*80)
print("CONCLUSIONES")
print("="*80)

# Comparar KNN
mejora_knn_r2 = (res_clean['knn']['r2'] - res_orig['knn']['r2']) / res_orig['knn']['r2'] * 100
mejora_knn_mae = (res_orig['knn']['mae'] - res_clean['knn']['mae']) / res_orig['knn']['mae'] * 100  # negativo si empeora
mejora_rf_r2 = (res_clean['rf']['r2'] - res_orig['rf']['r2']) / res_orig['rf']['r2'] * 100
mejora_rf_mae = (res_orig['rf']['mae'] - res_clean['rf']['mae']) / res_orig['rf']['mae'] * 100

print(f"\n🔷 KNN:")
print(f"   R²: {res_orig['knn']['r2']:.4f} \u2192 {res_clean['knn']['r2']:.4f} ({mejora_knn_r2:+.1f}%)")
print(f"   MAE: ${res_orig['knn']['mae']:,.0f} \u2192 ${res_clean['knn']['mae']:,.0f} ({mejora_knn_mae:+.1f}%)")

print(f"\n🔶 Random Forest:")
print(f"   R²: {res_orig['rf']['r2']:.4f} \u2192 {res_clean['rf']['r2']:.4f} ({mejora_rf_r2:+.1f}%)")
print(f"   MAE: ${res_orig['rf']['mae']:,.0f} \u2192 ${res_clean['rf']['mae']:,.0f} ({mejora_rf_mae:+.1f}%)")

print("\n\uFFFD INTERPRETACIÓN DE NEGOCIO:")
if mejora_knn_r2 > 1 and mejora_knn_mae > 1:
    print(" - KNN mejora al eliminar outliers (era sensible a ellos).")
elif mejora_knn_r2 < -1 and mejora_knn_mae < -1:
    print(" - KNN empeora al eliminar outliers (depende de ellos para generalizar).")
else:
    print(" - KNN es relativamente estable ante outliers en este dataset.")

if mejora_rf_r2 < -2 and mejora_rf_mae < -2:
    print(" - Random Forest empeora claramente al eliminar outliers, lo que indica que los casos extremos (locales premium) contienen información valiosa que el modelo aprovecha.")
elif mejora_rf_r2 > 1 and mejora_rf_mae > 1:
    print(" - Random Forest mejora sin outliers, lo que sugiere que algunos outliers eran ruido.")
else:
    print(" - Random Forest es robusto; los outliers no afectan significativamente su rendimiento.")

# Recomendación final
if res_orig['rf']['r2'] > res_clean['rf']['r2']:
    print("\n\u2705 RECOMENDACIÓN: Mantener los outliers en el dataset.")
    print("   Random Forest alcanza su mejor desempeño con ellos (R² original más alto).")
else:
    print("\n\u2705 RECOMENDACIÓN: Eliminar outliers antes de entrenar.")
    print("   El modelo mejora su precisión sin ellos.")
```
